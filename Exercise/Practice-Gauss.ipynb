```python
def checkDiagonal(x):
    x = np.array(x)
    diag = np.diag(np.abs(x))
    offDiag = np.sum(np.abs(x), axis = 1) - diag
    
#     for i in range(len(x)):
#         print(f"Diagonal {i+1} = {diag[i]}")
#         print(f"Total of not diagonal value = ", end="")
#         printed = False
#         for j, val in enumerate(x[i]):
#             if j != i:
#                 if printed != False:
#                     print("+", end=" ")
#                 print(f"|{val}|", end=" ")
#                 printed = True
#         print(f"= {offDiag[i]}")
#         if diag[i] > offDiag[i]:
#             print(f"{diag[i]} > {offDiag[i]}")
#         else:
#             print(f"Because {diag[i]} < {offDiag[i]}, The Equation is not diagonally dominant!")
#             return False
            
    if np.all(diag > offDiag):
        print("This Equation is Diagonally Dominant!")
        return True
    else:
        print("Not Diagonally Dominant")
        return False

for i, val in enumerate(mX):
    print(f"Equation {i+1} Array X:")
    checkDiagonal(val)
    print("\n")
```

```python
def gaussSeidel(x, y, epsilon, max):
    x = np.array(x)
    y = np.array(y)
    diag = np.diag(x)
    x = -x
    np.fill_diagonal(x, 0)
    xOld = np.zeros(x.shape[0])
    
    for i in range(len(x)):
        print(f"x{i} = ", end="")
        print(f"({y[i]} ", end="")
        printed = True
        for j, val in enumerate(x[i]):  
            if j != i:
                if printed != False:
                    print(" +", end = " ")
                print(f"({val}x{j+1})", end="")
                printed = True
        print(f" / {diag[i]}")
    
    for i in range(max):
        xNew = np.array(xOld)
        print(f"Iteration: {i+1}|", end=" ")
        coma = False
        for j, row in enumerate(x):
            xNew[j] = (y[j] + np.dot(row, xNew)) / diag[j]
            if coma != False:
                print(",", end=" ")
            print(f"x{j+1} = {xNew[j]:.8f}", end = "")
            coma = True
        print()
        distance = np.sqrt(np.dot(xNew - xOld, xNew - xOld))
        if distance < epsilon:
            print("\nConverged!")
            print("Result: ")
            for i, val in enumerate(x):
                print(f"x{i+1} = {xNew[i]:.8f}")
            return True
        xOld = xNew
    return False

for i, (x, y) in enumerate(zip(mX, mY)):
    print(f"Equation {i+1}")
    if checkDiagonal(x):
        if not gaussSeidel(x, y, 0.01456, 5):
            print("Not Converged! need more iterations")
    print("\n=============================")
```
