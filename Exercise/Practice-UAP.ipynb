```python
def checkDiagonal(x):
    x = np.array(x)
    diag = np.diag(np.abs(x))
    offDiag = np.sum(np.abs(x), axis=1) - diag
    
    if np.all(diag > offDiag):
        return True
    else:
        return False
    
def gaussSeidel(x, y, epsilon, max):
    x = np.array(x)
    y = np.array(y)
    diag = np.diag(x)
    x = -x
    np.fill_diagonal(x, 0)
    xOld = np.zeros(x.shape[0])
    
    for i in range(max):
        xNew = np.array(xOld)
        for j, row in enumerate(x):
            xNew[j] = (y[j] + np.dot(row, xNew)) / diag[j]
        
        print(f"iter: {i} {xNew}")
            
        distance = np.sqrt(np.dot(xNew - xOld, xNew - xOld))
#         print(distance)
        if distance < epsilon:
            print("convergent")
            return True
        xOld = xNew
    return False

for i, (x, y) in enumerate(zip(Xs, Ys)):
    print(f"A: {x} y: {y}")
    if checkDiagonal(x):
        if not gaussSeidel(x, y, 0.01, 20):
            print("not convergent")
    else:
        print("not diagonally dominant")
    print()
```

```python
def newtonRhapson(f, df, x0, tolerance, max, i):
    if max == 0:
        print("max iteration reached!")
        return
    if np.abs(f(x0)) < tolerance:
        print(f"\nThe root of equation is {x0:.6f}")
        return
    else:
        xNew = x0 - (f(x0)/df(x0))
        print(f"Iteration-{i}, new root = {xNew:.6f}")
        newtonRhapson(f, df, xNew, tolerance, max-1, i+1)
    

x = symbols("x")
f = lambda x: 4*x**3 + 2
df = diff(f(x), x)
df = lambdify(x, df, "numpy")
newtonRhapson(f, df, 3, 0.001, 30, 1)
```
